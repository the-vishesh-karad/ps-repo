```python
import pandas as pd

# Read the dataset (use the uploaded file path)
data = pd.read_csv('/content/sales_data_sample.csv', encoding='latin1')

# Display first few rows
data.head()
```

```python
# Check summary statistics
data.describe()

```

```python
# Check missing values
data.isnull().sum()
```

```python
# Drop rows with too many missing values (optional)
data = data.dropna(thresh=len(data.columns) - 3)
```

```python
# Fill missing values if necessary
data = data.fillna("Not Available")
```

```python
# Check for duplicates
data.duplicated().sum()
```

```python
# Remove duplicates
data = data.drop_duplicates()
```

```python
# Check total sales by product line
data.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)
```

```python
# Check top 5 countries by total sales
data.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False).head(5)
```

```python
# Average sales per deal size
data.groupby('DEALSIZE')['SALES'].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1️⃣ Sales by Product Line
plt.figure(figsize=(10,5))
sns.barplot(x='PRODUCTLINE', y='SALES', data=data, estimator=sum, ci=None)
plt.xticks(rotation=45)
plt.title("Total Sales by Product Line")
plt.show()
```

```python
# 2️⃣ Sales by Country
plt.figure(figsize=(12,6))
top_countries = data.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False).head(10)
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title("Top 10 Countries by Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3️⃣ Trend of Sales by Year
plt.figure(figsize=(10,5))
data.groupby('YEAR_ID')['SALES'].sum().plot(kind='line', marker='o')
plt.title("Sales Trend by Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(data[['SALES', 'QUANTITYORDERED', 'PRICEEACH']])
plt.suptitle("Pairplot of Sales Metrics", y=1.02)
plt.show()
```

```python
X = data[['QUANTITYORDERED', 'SALES']].dropna()

kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(X)

plt.figure(figsize=(10,6))
sns.scatterplot(x='QUANTITYORDERED', y='SALES', hue='Cluster', data=data, palette='viridis')
plt.title("K-Means Clustering of Sales Data")
plt.show()
```

```python
country_sales = data.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=country_sales.index, y=country_sales.values, palette='cool')
plt.xticks(rotation=45)
plt.title("Top 10 Countries by Total Sales")
plt.ylabel("Total Sales")
plt.xlabel("Country")
plt.show()
```
